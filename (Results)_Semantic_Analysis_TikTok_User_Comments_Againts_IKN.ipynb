```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'df' with the 'label' column
label_counts = df['label'].value_counts()

# Map label values to corresponding labels
label_names = {1: 'Komentar Positif', 0: 'Komentar Negatif'}
label_names = [label_names[label] for label in label_counts.index]

# Create a bar plot with modified x-axis labels and annotations
bars = plt.bar(label_names, label_counts.values)
plt.xlabel('Label')
plt.ylabel('Jumlah')
plt.title('Distribusi Label')

# Add count values as annotations on top of each bar
plt.bar_label(bars, label_counts.values, label_type='edge')

plt.show()
```

**Exp:** Diperoleh 2898 users yang berkomentar terhadapa isu pemindahan IKN, 1185 diantaranya merupakan komentar positif dan 1713 merupakan komentar negatif
