### Turn list into string and lowercase

### Remove Emoji

```python
def remove_emojis(text):
    emoji = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642"
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoji, '', text)

combined_df['Comment Text'] = combined_df['Comment Text'].apply(remove_emojis)
combined_df
```

### Remove Punctuation

```python
punctuation =   "!\"#$%&'()*+,-./:;<=>?[\]^_`{|}~‚Äº·•¨·≠Ñ‡ºä"  #string.punctuation
combined_df['Comment Text'] = combined_df['Comment Text'].apply(lambda x: ''.join([c for c in x if c not in punctuation]))
translator = str.maketrans('', '', punctuation)
combined_df['Comment Text'] = combined_df['Comment Text'].str.translate(translator)
combined_df
```

```python
combined_df['Comment Text'] = combined_df['Comment Text'].str.replace('@', '')
```

```python
def handle_typos(text):
    words = text.split()
    normalized_words = []
    for word in words:
        if word.lower() in typos_all and len(word.split()) == 1:
            normalized_words.append(typos_all[word.lower()])
        else:
            normalized_words.append(word)
    return ' '.join(normalized_words)
```

### Remove Stopwords

```python
import nltk
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stopwords_id = stopwords.words('indonesian')
clean_comments = []
for comment in combined_df['Comment Text']:
    words = comment.split()
    clean_comment = [word for word in words if word.lower() not in stopwords_id]
    clean_comment = ' '.join(clean_comment)
    clean_comments.append(clean_comment)
combined_df['Comment Text'] = clean_comments
```
